```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import os
from pydub import AudioSegment

startTime = 10*1000
endTime = -10*1000
endTimeInSeconds = 10
segmentSize = 30*1000

rootdir = "audio"
for root, dirs, files in os.walk(rootdir):
    for file in files:
        if file.endswith(".mp3"):
            fileName = os.path.join(root, file)
            if fileName.endswith("-sliced.mp3"):
                continue
            print("Processing " + fileName)
            if os.path.isfile(fileName.replace(".mp3", "-1-sliced.mp3")):
                #This file has already been processed in a previous iteration, so skip it
                print("Already processed")
                continue
            try:
                song = AudioSegment.from_file(fileName, format="mp3", start_second = 10)
            except:
                print("Couldn't decode " + fileName)
                continue
            print("Duration is " + str(len(song)))
            slices = list(song[::segmentSize])
            print("Number of slices is " + str(len(slices)))
            if (slices[len(slices) - 1].duration_seconds < endTimeInSeconds):
                slices.pop()
            for i, chunk in enumerate(slices):
                #print("Writing slice " + str(i) + " of " + str(len(slices)))
                extractedFileName = fileName.replace(".mp3", "-" + str(i) + "-sliced.mp3")
                # If it is the last slice, then truncate 10 seconds from the end.
                if (i == len(slices)-1):
                    #print("Truncating last segment")
                    chunk = chunk[:endTime]
                chunk.export(extractedFileName, format="mp3")
                #print("Done")
                
```
